```python
import numpy as np
import pandas as pd

df = pd.read_csv('https://data.ssb.no/api/v0/dataset/1122.csv?lang=en', sep=",")
df
```

```python
df_copy = df.copy()
```

```python
df_copy['week'] = df_copy['week'].apply(lambda x: x.replace('U','-W'))

#df_copy['week'] = pd.to_datetime(df_copy['week'])

df_copy
```

```python
# you are having issues with week numbers in POWER BI. Python do not understand that 2004 had 53 weeks. 
# It labels 2004-W53 and 2005-W01 as the same


df_copy.iloc[520:,:].head(20)
```

```python
import datetime

# for row in df_copy['week']:
#     row = datetime.datetime.strptime(row + '-1', "%Y-W%W-%w")

# df_copy['week'] = datetime.datetime.strptime(df_copy['week'] + '-1', "%Y-W%W-%w")

df_copy['week'] = df_copy['week'].apply(lambda x: datetime.datetime.strptime(x + '-1', "%Y-W%W-%w"))

df_copy
```

```python
# fresh fish 

df_fresh = df_copy[df_copy['commodity group']=='01 Fish-farm bred salmon, fresh or chilled']

df_frozen = df_copy[df_copy['commodity group']=='02 Fish-farm bred salmon, frozen']

df_fresh.drop('commodity group',axis=1, inplace = True)

df_frozen.reset_index(drop=True,inplace=True)

df_frozen.drop('commodity group',axis=1, inplace = True)
df_frozen
```

```python
weeks = df_copy['week'][::2]

weeks.reset_index(drop=True, inplace=True)

weeks
```

```python
df_frozen['price_kg']=0

df_frozen = pd.DataFrame(df_frozen['03024: Export of salmon, fish-farm bred, by commodity group, week and contents'].values.reshape(-1, 2), columns = ['03024: Export of salmon, fish-farm bred, by commodity group, week and contents', 'price_kg'])

df_frozen.columns=['Weight_Tons', 'Price_kg']

df_frozen = pd.merge(df_frozen, pd.DataFrame(weeks), left_index=True, right_index=True)

df_frozen
```

```python
df_fresh['price_kg']=0

df_fresh = pd.DataFrame(df_fresh['03024: Export of salmon, fish-farm bred, by commodity group, week and contents'].values.reshape(-1, 2), columns = ['03024: Export of salmon, fish-farm bred, by commodity group, week and contents', 'price_kg'])

df_fresh.columns=['Weight_Tons', 'Price_kg']

df_fresh = pd.merge(df_fresh, pd.DataFrame(weeks), left_index=True, right_index=True)

df_fresh
```

```python
df_full = pd.merge(df_fresh, df_frozen, left_index=True, right_index=True)

df_full.columns=['Weight_Ton_Fresh', 'Price_kg_Fresh', 'Week', 'Weight_Ton_Frozen', 'Price_kg_Frozen', 'week_y']

df_full.drop('week_y', axis=1,inplace=True)

df_full[['Week','Weight_Ton_Fresh', 'Price_kg_Fresh',  'Weight_Ton_Frozen', 'Price_kg_Frozen']]
```

```python
df_full.to_excel('fish_export_clean.xlsx', index=False)
```
